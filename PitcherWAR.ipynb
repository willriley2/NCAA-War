```python
adjustment = lg_runs_allowed_per_9 - lg_era
```

```python
df['fip_runs_per_9_scale'] = df['FIP'] + adjustment
```

```python
#set constant for league park factor, at 1.05
#socon = boom box

lg_park_factor = 105
```

```python
df['park_adjusted_fip_runs_per_9_scale'] = df['fip_runs_per_9_scale'] / (lg_park_factor/100)
```

```python
#compare to league fip
dfLg['FIP'] = ((13*dfLg['HR']) + (3*dfLg['BB'] + dfLg['HBP']) - (2*dfLg['SO'])) / (dfLg['IP']) + fip_constant
dfLg['FIP']
```

```python
lg_fip = 5.426844
```

```python
df['runs_above_average_per_9'] = lg_fip - df['park_adjusted_fip_runs_per_9_scale']
```

```python
df['IP/G'] = df['IP'] / df['APP']
```

```python
df['dynamic_runs_per_win'] = (((18 - df['IP/G'])*(lg_fip) + ((df['IP/G']*df['park_adjusted_fip_runs_per_9_scale']))/18)+2) * 1.5
```

```python
df['wins_per_game_above_average'] = df['runs_above_average_per_9'] / df['dynamic_runs_per_win']
```

```python
df['replacement_level'] = (0.03*(1-(df['APP']/df['TeamGames']))) + (0.12*(1-(df['GS']/df['TeamGames'])))
```

```python
df['wins_per_game_above_replacement'] = df['wins_per_game_above_average'] + df['replacement_level']
```

```python
df['WAR'] = df['wins_per_game_above_replacement'] * (df['IP']/9)
```

```python
dfWAR = df[['Player', 'Team', 'WAR']]
```

```python
dfWAR
```

```python
dfWAR['war_per_game'] = dfWAR['WAR'] / df['APP']
```

```python
dfWAR['war_per_56'] = dfWAR['war_per_game'] * 56
```
